## Cleaning Data

### Cleaning Wrong Format Data

```python
import re

def random_month_year(year:str):
    year_abbr = str(year)[-2:]
    return str("20"+year_abbr)

def is_month_year_format(date:str):
    pattern = r'^[A-Za-z]{3}-\d{2}$'
    return bool(re.match(pattern, date))

def correct_format(data:str):
    if is_month_year_format(data) :
        return random_month_year(data)
    return data

# df.loc[0,'registration_year'] 

for index, value in df['registration_year'].items():
    df.loc[index,'registration_year'] = correct_format(value)


df["registration_year"].sort_values(ascending=True)

# df.info()
```

### Handle Outliers

```python
def outlier_handle(data:pd.Series,lower = False):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    
    upper_bound = Q3 + 1.5 * IQR
    print(data.name)
    print("\tUpper Bound :",upper_bound)

    if lower:
        lower_bound = Q1 - 1.5 * IQR
        print("\tLower Bound :",lower_bound)
        return data[(data > upper_bound) | (data < lower_bound)]
    
    return data[(data > upper_bound)]
```

คุณภาพ หรือขอ้ มูลเชิงปริมาณที่แบ่งเป็นช่วง และขอ้ มูลไม่จา เป็นตอ้งมีการแจกแจงแบบปกติตัว

t-test กับ ANOVA ต่างกันอย่างไร?
t-test และ ANOVA (Analysis of Variance) เป็นเครื่องมือที่ใช้ในการเปรียบเทียบค่าเฉลี่ยในชุดข้อมูลต่างๆ อย่างไรก็ตาม มีความแตกต่างที่สําคัญบางประการระหว่างพวกเขา:

จํานวนกลุ่ม:
การทดสอบ T: ใช้เมื่อเปรียบเทียบค่าเฉลี่ยของข้อมูลสองชุด
อโนวา: ใช้เมื่อเปรียบเทียบค่าเฉลี่ยของชุดข้อมูลตั้งแต่สามชุดขึ้นไป

ประเภทของตัวแปร:
การทดสอบ T: ทํางานร่วมกับตัวแปรตัวเลขต่อเนื่องและข้อมูลอิสระ
อโนวา: ทํางานร่วมกับตัวแปรตัวเลขอย่างต่อเนื่องและสามารถจัดการทั้งข้อมูลอ้างอิงและอิสระ

ประเภทของผลลัพธ์:
การทดสอบ T: ให้ค่า t ซึ่งแสดงให้เห็นว่าความแตกต่างเฉลี่ยระหว่างสองกลุ่มมีนัยสําคัญเพียงใด
อโนวา: ให้ค่า F ซึ่งระบุถึงความสําคัญของความแตกต่างของค่าเฉลี่ยระหว่างสามกลุ่มขึ้นไป

ประเภทของการวิเคราะห์:
การทดสอบ T: ดําเนินการวิเคราะห์ตัวแปรเดียวตรวจสอบตัวแปรอิสระทีละตัว
อโนวา: ดําเนินการวิเคราะห์หลายตัวแปรช่วยให้สามารถตรวจสอบปัจจัยอิสระหลายอย่างพร้อมกันได้

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("clean_used_car.csv")
```

```python
def oneway_anova(column,dataframe = df):
    grouped_data = [group['price(in lakhs)'].values for name, group in dataframe.groupby(column, observed=True)]
    f_statistic, p_value = stats.f_oneway(*grouped_data)

    print("Count Data:",len(grouped_data))
    print("F-statistic:", f_statistic)
    print("p-value:", p_value)

def barplot_price(column , xsize = 10 , ysize = 5 , dataframe = df):
    mean_prices = dataframe.groupby(column, observed=True)['price(in lakhs)'].mean().reset_index()
    mean_prices = mean_prices.sort_values("price(in lakhs)",ascending=False)
    plt.figure(figsize=(xsize, ysize))
    sns.barplot(x=column, y='price(in lakhs)', data=mean_prices, color='skyblue')
    plt.xlabel(column)
    plt.ylabel('Average Price (in lakhs)')
    plt.title('Average Price per '+column)
    plt.show()
```
